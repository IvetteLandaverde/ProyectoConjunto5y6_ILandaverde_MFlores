### Solución Moisés:

Para obtener la función inversa que distribuya como la función de distribución acumuldada, calculamos:

$$F(x)=\frac{2}{\pi}\sin^{-1}(\sqrt x)= U,   U \sim \text{u[0,1]}$$

$$\sin^{-1}(\sqrt x)= \frac{\pi U}{2}$$

$$\sin(\sin^{-1}(\sqrt x))= \sin(\frac{\pi U}{2})$$

$$\sqrt x = \sin(\frac{\pi U}{2})$$

$$x = \sin^{2}(\frac{\pi U}{2})$$

Ahora, para obtener la función de distribución $f(x)$ se requiere deriva $F(x)$:

$$\frac{dF(x)}{dx} = f(x)$$

$$\frac{d}{dx}(\frac{2}{\pi}\sin^{-1}(\sqrt x))=\frac{2}{\pi} \frac{d}{dx}(\sin^{-1}(\sqrt x))$$

$$\frac{d}{dx}(\frac{2}{\pi}\sin^{-1}(\sqrt x))=\frac{2}{\pi} \frac{1}{\sqrt{1-\sqrt{x^{2}}}} \frac{1}{2\sqrt{x}} = \frac{1}{\pi} \frac{1}{\sqrt{1-x}} \frac{1}{\sqrt{x}}$$

$$\frac{d}{dx}(\frac{2}{\pi}\sin^{-1}(\sqrt x))=\frac{1}{\pi\sqrt{1-x}\sqrt{x}}$$

```python
### Está función retorna la función de distribución acumulada
def D_browniano(N:"Número de muestras"):
    "Inversa de la función de distribución F(x)=\frac{2}{\pi}\sin^{-1}(\sqrt x),\quad 0\leq x\leq 1$"
    return np.sin((np.random.rand(N)*np.pi)/2)**2###Hacemos la función inversa
f_x = lambda x: 1/(np.pi*np.sqrt(1-x)*np.sqrt(x))###Función de distribución f(x) obtenida de la derivada planteada
```

> 2. Generación variable aleatoria Discreta
> La distribución binomial modela el número de éxitos de n ensayos independientes donde hay una probabilidad p de éxito en cada ensayo.
> Generar una variable aletoria binomial con parámetros $n=10$ y $p=0.7$. Recordar que $$X\sim binomial(n,p) \longrightarrow p_i=P(X=i)=\frac{n!}{i!(n-i)!}p^i(1-p)^{n-i},\quad i=0,1,\cdots,n$$
> Por propiedades de la operación factorial la anterior $p_i$ se puede escribir como:
> $$p_{i+1}=\frac{n-i}{i+1}\frac{p}{1-p} p_i $$

### Solución  Ivette: 

```python
N=100
def binom():
    x = 10; p = 0.7; N = 100
    result = []
    for a in range(N):
        result.append(scipy.stats.binom.cdf(x, N, p))
        x += 3
    return result
def Gen_distr_discreta(p_acum: 'P.Acumulada de la distribución a generar',
                       N: 'cantidad de números aleatorios a generar'):
    
    U =np.random.rand(N)
    # los valores aletorios
    y = ([sum([1 for p in p_acum if p < ui]) for ui in U])
    return y
def plot_histogram_discrete(distribucion:'distribución a graficar histograma',
                            label:'label del legend'):
    # len(set(distribucion)) cuenta la cantidad de elementos distintos de la variable 'distribucion'
    plt.figure(figsize=[6,4])
    y,x = np.histogram(distribucion,bins = len(set(distribucion))) 
    plt.bar(list(set(distribucion)),y,label=label,color="m",linestyle='dashed')
    plt.legend()
    plt.grid()
    plt.show()
p_acum = binom() 
distribucion = Gen_distr_discreta(p_acum,N) 
plot_histogram_discrete(distribucion,"Distribución de función")

```

### Solución  Moisés: 

```python
###Se hace una función de distribución acumulada
def d_binomial(n:"Parámetro n",p:"Parametro p"):
    "Está función genera la distribución binomial acumulada"
    vector = np.zeros(n) ###Se generá un vector de n zeros
    vector[0] = (1-p)**n ###Se llena la posición cero 
    def llenar_vector(i): 
        nonlocal vector 
        vector[i+1] = ((n-i)/(i+1))*((p)/(1-p))*vector[i]  ###Fórmula del factorial para llenar el vector

    [llenar_vector(i) for i in range(n-1)] ###Compresión de listas para llenar el vector
    return np.cumsum(vector)### Se regresa el acumulado al ser función de distribución acumulada
```

```python
###Se usa la función vista en clase para generar la distribución discreta
def Gen_distr_discreta(p_acum: 'P.Acumulada de la distribución a generar',
                       N: 'cantidad de números aleatorios a generar'):
    "Genera una distribución discreta a partir de una distribución acumulada"
    U =np.random.rand(N)
    v = [sum([1 for p in p_acum if p < ui ])  for ui in U]
    return v
###Se utiliza la formula vista en clase para plotear la distribución discreta generada
def plot_histogram_discrete(distribucion:'distribución a graficar histograma',
                            label:'label del legend'):
    plt.figure(figsize=[8,4])
    y,x = np.histogram(distribucion,bins = len(set(distribucion))) 
    plt.bar(list(set(distribucion)),y,label=label)
    plt.legend()
    plt.show()
```

```python
n,p = 10 ,0.7###Declaramos los parametros n y p
p_acum = d_binomial(n,p) ###Se guarda en una variable la distribución aucmulada
N = 1000 ###El número de valores aleatorios a generar dentro de nuestra distribución discreta
distribucion = Gen_distr_discreta(p_acum,N) ###Se guarda la distribución discreta generada en una variable
plt.figure(figsize=[8,6])
plot_histogram_discrete(distribucion,"Distribución con función propia")###Ploteamos 
```
