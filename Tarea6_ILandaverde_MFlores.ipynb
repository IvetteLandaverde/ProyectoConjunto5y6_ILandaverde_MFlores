### Solución  Ivette: 

Para encontrar la función de densidad g(x) recurrimos a las integrales por lo que se integra la función $t(x)$ que es $a \sin(\pi x)$ para conocer c:
> Integral:
$$\int_{0}^{1} t(x) dx = \int_{0}^{1} a \sin(\pi x) dx$$
> $a$ se toma como constante:
$$a \int_{0}^{1}  \sin(\pi x) dx$$
> ahora se aplica integración por sustitución siendo $u = \pi x$
> Entonces:
$$\frac{a}{\pi} \int_{0}^{\pi}  \sin(u) du$$
> Lo cual da como resultado: $$\frac{a}{\pi} [-\cos(u)]_{0}^{\pi}$$
$$ c= \frac{2a}{\pi} $$
> Por lo que para conocer g(x) sólo se divide $t(x)$ entre $c$:
$$\frac{t(x)}{c} = \frac{a \sin(\pi x)}{\frac{2a}{\pi}} = \frac{\pi \sin(\pi x)}{2}$$
<span style="color:red"> Por lo que: $$g(x) = \frac{\pi \sin(\pi x)}{2}$$

### Solución Moisés:

### Respuesta planteada en el ejercicio es:
$g(x)$ está definida como $g(x) = \frac{t(x)}{c}$. Tenemos que $t(x)=a \sin(\pi x)$ y $c$ como $c =\int_{0}^{1} t(x) dx$. Entonces se pude operar de la siguiente manera:

$$ c =\int_{0}^{1} t(x) dx =\int_{0}^{1} a \sin(\pi x) dx  $$
$$ a\int_{0}^{1} \sin(\pi x) dx = a[\frac{-\cos(\pi x)}{\pi}]_{0}^{1} $$
$$ a[\frac{-\cos(\pi x)}{\pi}]_{0}^{1} = \frac{a}{\pi} - (- \frac{a}{\pi}) = \frac{2a}{\pi} $$
$$ c = \frac{2a}{\pi} $$

Ahora, simplificamos $g(x) = \frac{t(x)}{c}$

$$g(x) = \frac{a \sin(\pi x)}{\frac{2a}{\pi}} = \frac{\frac{a \sin(\pi x)}{1}}{\frac{2a}{\pi}} = \frac{a\pi \sin(\pi x)}{2a} = \frac{\pi \sin(\pi x)}{2}$$

Obtenemos: 
$$ g(x) = \frac{\pi \sin(\pi x)}{2}$$

### <font color ='blue'> **Ejercicio 8:**  
Demostrar la validez de la siguiente ecuación si 
>$$X\sim binomial(n,p) \longrightarrow p_i=P(X=i)=\frac{n!}{i!(n-i)!}p^i(1-p)^{n-i},\quad i=0,1,\cdots,n$$
>$$p_{i+1}=\frac{n-i}{i+1}\frac{p}{1-p} p_i \longrightarrow \text{Hablar de las ventajas que sea recursiva}$$

### Solución  Ivette: 

> Se sabe que:
$$p_{i+1}=\frac{n-i}{i+1}\frac{p}{1-p} p_i $$
> por lo que despejando:
$$\frac{p_{i+1}}{p_i}=\frac{n-i}{i+1}\frac{p}{1-p} $$
> Y por propieddes factoriales puedo sustituir:
$$\frac{(n+1)!}{n!}=n+1 ; \frac{n!}{(n+1)!}=\frac{1}{n+1}$$
> y:
$$ (n-1)! = \frac{n!}{n} ; n!=n(1+n)!$$
> Y tomando en cuenta que:
>$$X\sim binomial(n,p) \longrightarrow p_i=P(X=i)=\frac{n!}{i!(n-i)!}p^i(1-p)^{n-i},\quad i=0,1,\cdots,n$$
> Entonces simplificando obtenemos::
$$\frac{p_{i+1}}{p_i} = \frac{p(n-i)!(-p+i)(i+1)^{n+1-i}}{(i+1)(n+1-i)(-p+1)^{n-i}}$$

$$p_{i+1} = \frac{p(n-i)!(-p+i)(i+1)^{n+1-i}}{(i+1)(n+1-i)(-p+1)^{n-i}}p_i$$

$$p_{i+1} = \frac{n-i}{i+1}\frac{p}{1-p}p_i$$


<font color ='blue'>La ventaja de que sea recursiva es que es menos compleja porque nos ahorraríamos toda la parte de los factoriales por lo cual es más fácil de programar y necesita menos memoria computacional.
    
    

### Solución Moisés:

### Respuesta planteada en el ejercicio es:
Si se desea demostrar matemaicamente que: $p_{i+1}=\frac{n-i}{i+1}\frac{p}{1-p} p_i$, primeramente definimos las siguientes propiedades de los factoriales: $$1. n!=n(1+n)!$$ y $$2. \frac{n!}{(n+1)!}=\frac{1}{n+1}$$

Se tiene que,
$$ \frac{p_{i+1}}{p_i}=\frac{\frac{n! p ^ {i+1} (1-p)^{n-i+1}}{(i+1)!(n-i+1)!}}{\frac{n!}{i!(n-i)!}p^i(1-p)^{n-i}}$$
    
$$\frac{p_{i+1}}{p_i} = \frac{n!p^ip(1-p)^ni!(n-i)!(1-p)^i}{(1-p)^i(1-p)(i+1)!(n-i-1)!n!p^i(1-p)^n}$$
    
$$\frac{p_{i+1}}{p_i} = \frac{pp^i(1-p)^ni!(n-i)!(1-p)^i}{(1-p)^i(1-p)(i+1)!(n-i-1)!p^i(1-p)^n} = \frac{pi!(n-i)!(1-p)^i}{(1-p)^i(1-p)(i+1)!(n-i-1)!}=\frac{pi!(n-i)!}{(1-p)(i+1)!(n-i-1)!}$$ 

$$\frac{p_{i+1}}{p_i} = \frac{pi!(n-i)!}{(1-p)i!(i+1)(n-i-1)!} = \frac{p(n-i)!}{(1-p)(i+1)(n-i-1)!}$$
    
$$\frac{p_{i+1}}{p_i} = \frac{p(n-i)}{(1-p)(i+1)} = \frac{n-i}{i+1}\frac{p}{1-p}$$
    
$$p_{i+1} = \frac{n-i}{i+1}\frac{p}{1-p}p_i$$
    
Las ventajas de que sea recursiva nos ayudan a programar a través de vectores el comportamiento de dicha sucesión, ya que de no contar con está recursividad se vuelve compleja la programación de los factoriales. Así mismo, cuando tenemos un comportamiento recursivo podemos manejar más facilmente lor resultados que se van generando y usarlos de acuerdo a nuestros propósitos. Entre otras cosas nos permite pensar en la optimización, cómo ya se mencionó, por medio de la programación vectorizada.
    
Para obtener el valor inicial y aplicarlo en la sucesión se tiene que si $i=0$, entonces:
    
$$P(X=0)=\frac{n!}{0!(n-0)!}p^0(1-p)^{n-0} = \frac{n!}{1(n)!}1(1-p)^{n}$$ 
    
$$P(X=0)=\frac{n!}{1(n)!}1(1-p)^{n}=\frac{n!}{(n)!}(1-p)^{n}=1(1-p)^{n}$$
    
$$P(X=0)=(1-p)^{n}$$

```python
def Gen_distr_discreta(p_acum: 'P.Acumulada de la distribución a generar',
                       indices: 'valores reales a generar aleatoriamente',
                       N: 'cantidad de números aleatorios a generar'):
    
    U =np.random.rand(N)
    # Diccionario de valores aleatorios
    rand2reales = {i: idx for i, idx in enumerate(indices)}

    # Series de los valores aletorios
    y = pd.Series([sum([1 for p in p_acum if p < ui]) for ui in U]).map(rand2reales)

    return y
def plot_histogram_discrete(distribucion:'señal de varibles aleatorias de un distribución DISCRETA dada',
                            label:'label del legend a aparecer en el gráfica',
                            densidad:'por defecto regresa el histograma en densidad'=True):
    # len(set(distribucion)) cuenta la cantidad de elementos distintos de la variable 'distribucion'
    plt.figure(figsize=[8,4])
    y, x = np.histogram(distribucion, bins=len(set(distribucion)), density=densidad) 
    plt.bar(x[1:], y, label=label)
    plt.legend()
    plt.grid()
    plt.title('Poisson',fontsize=14)
    plt.xlabel('Rango',fontsize=14)
    plt.ylabel('Probabilidad',fontsize=14)
    plt.show()  
# Función para la distribución acumulada de Poisson
#hecha como la P_acum_Binomial_o de la clase 10:
def poisson(l,k):
    z=np.zeros(k+1)
    z[0]=np.exp(-l)
    def z1(i):
        nonlocal z
        z[i+1]=z[i]*(l/(i+1))
    # Lleno el vector Pr usando compresión de listas
    [z1(i) for i in range(k-1)]
    return np.cumsum(z)
N = 10000;k = 10
Poisson=poisson(5,5)
# Usando el método de la transformada inversa
pl= Gen_distr_discreta(Poisson,np.arange(0,k+1),N)    
plot_histogram_discrete(pl,"Distribución Poisson")  
```

Si se desea demostrar matemaicamente que: $P(k+1)=\frac{\lambda}{k+1}P(k)$, primeramente definimos las siguientes propiedades de los factoriales: $$1. k!=k(1+k)!$$ y $$2. \frac{k!}{(k+1)!}=\frac{1}{k+1}$$


Se tiene que 
$$\frac{P(k+1)}{P(k)} = \frac{\frac{e^{-\lambda}(\lambda)^{k+1}}{(k+1)!}}{\frac{e^{-\lambda}(\lambda)^k}{k!}}$$

$$\frac{P(k+1)}{P(k)} = \frac{e^{-\lambda}\lambda^{k+1} k!}{(k+1)!e^{-\lambda}\lambda^k} = \frac{e^{-\lambda}\lambda^{k}\lambda k!}{(k+1)!e^{-\lambda}\lambda^k}$$

$$\frac{P(k+1)}{P(k)} = \frac{\lambda k!}{(k+1)!}$$

$$\frac{P(k+1)}{P(k)} = \frac{k!}{(k+1)!} * \frac{\lambda}{1}$$

$$\frac{P(k+1)}{P(k)} = \frac{k!}{(k+1)!} * \frac{\lambda}{1} = \frac{1}{(k+1)} * \frac{\lambda}{1}=\frac{\lambda}{(k+1)}$$

$$P(k+1) = \frac{\lambda}{(k+1)}P(k)$$

Para obtener el valor inicial y aplicarlo en la sucesión se tiene que si $k=0$, entonces:
$$P(0,\lambda)=\frac{e^{-\lambda}(\lambda)^0}{0!}$$

$$P(0,\lambda)=\frac{e^{-\lambda}(1)}{1}=e^{-\lambda}$$

```python
###Se crea una función que genera la función de distribución de Poisson acumulada
def dist_acum_poisson(lam:"Parámetro lamda",n:"Término n"): 
    "Está función genera la distribución de Poisson Acumulada"
    vector = np.zeros(n) ###Se incia un vector de ceros hasta el término n
    vector[0] = (np.exp(-lam)) ###Se llena la primera posición del vector con el término calculado
    def llenar_vector(i):
        nonlocal vector
        vector[i+1] = (lam*vector[i])/(i+1) ###Fórmula iterativa para llenar el vector
    [llenar_vector(i) for i in range(n-1)] ###Comprensión delistas con la que se llena el vector
    return vector.cumsum() ###Se retorna de manera acumulada 
```

```python
### Fuunción para generar una distribución discreta
def Gen_distr_discreta(p_acum: 'P.Acumulada de la distribución a generar',
                       N: 'cantidad de números aleatorios a generar'):
    
    U =np.random.rand(N)
    v = [sum([1 for p in p_acum if p < ui ])  for ui in U]
    return v
### Función para plotear el histograma de la función de distribución
def plot_histogram_discrete(distribucion:'distribución a graficar histograma',
                            label:'label del legend'):
    plt.figure(figsize=[8,4])
    y,x = np.histogram(distribucion,bins = len(set(distribucion))) 
    plt.bar(list(set(distribucion)),y,label=label)
    plt.legend()
    plt.show()
    
n,lam= 1000,4 ###Se definen los parámetros n y lambda 
p_acum = dist_acum_poisson(lam,n) ###Se guardan en una variable la distribución de poisson
N=1000 ###El número de valores aleatorios a generar en la distribución discreta
distribucion = Gen_distr_discreta(p_acum,N) ###Se guarda la distribución discreta generada
plot_histogram_discrete(distribucion,"Distribución de Poisson") ###Se plotea el histograma
```
